```python
import pandas as pd

df = pd.read_csv('G:\.shortcut-targets-by-id\\1GLhesaekxPRKr-lkNYX-mtfouqJhqhXN\SIBEUX\KULIAH\\00 - UNEJ\Algorithm Programming\PEMROGRAMAN\Main Program\\utbk.csv')

lolos = 8
tidak_lolos = 22

def str(dx):
    suppdx1 = []
    suppdx2 = []
    lls = []
    tlls = []
    grx = []
    conf = []
    clulus = []
    ctlulus = []
    itemSet = df[dx].unique()

    for x in itemSet:
        ls = df[(df[dx]==x) & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]==x) & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        supls = len(ls)/lolos
        suptls = len(tls)/tidak_lolos
        growtValueMax = max(supls,suptls)
        growtValueMin = min(supls,suptls)
        GrowthRateValue = growtValueMax/growtValueMin
        grx.append(GrowthRateValue)
        conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        clulus.append(8)
        ctlulus.append(22)

    emerg = {'Range':itemSet[:],
            '|X| lulus':clulus[:],
            'Count Dx lulus': lls[:],
            '|X| tidak lulus':ctlulus[:],
            'Count Dx tidak lulus': tlls[:],
            'Supp D1(X)' : suppdx1[:],
            'Supp D2(X)' : suppdx2[:],
            'Growth Rate' : grx[:],
            'Confidence' : conf[:]}
    tabel = pd.DataFrame(emerg)
    display(tabel)


def num(dx,n=5,suppdx1=[],suppdx2=[],lls=[],tlls=[],grx=[],conf=[]):
    emerg = {'Range':['<400','401-500','501-600','601-700','>700'],
            '|X| lulus':[8,8,8,8,8],
            'Count Dx lulus': lls[:],
            '|X| tidak lulus':[22,22,22,22,22],
            'Count Dx tidak lulus': tlls[:],
            'Supp D1(X)' : suppdx1[:],
            'Supp D2(X)' : suppdx2[:],
            'Growth Rate' : grx[:],
            'Confidence' : conf[:]}
    if n==0:
        tabel = pd.DataFrame(emerg)
        display(tabel)

    # <400
    if n == 5:
        ls = df[(df[dx]<400) & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]<400) & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        if len(tls) == 0 or len(ls) == 0:
            grx.append(max(len(ls)/lolos,len(tls)/tidak_lolos))
            conf.append(max(len(ls)/lolos,len(tls)/tidak_lolos))

        else:
            supls = len(ls)/lolos
            suptls = len(tls)/tidak_lolos
            growtValueMax = max(supls,suptls)
            growtValueMin = min(supls,suptls)
            GrowthRateValue = growtValueMax/growtValueMin
            grx.append(GrowthRateValue)
            conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        num(dx,n-1,suppdx1,suppdx2,lls,tlls,grx,conf)
    # 401-500
    elif n == 4:
        ls = df[(df[dx]>400) & (df[dx]<500)  & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]>400) & (df[dx]<500)  & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        if len(tls) == 0 or len(ls) == 0:
            grx.append(max(len(ls)/lolos,len(tls)/tidak_lolos))
            conf.append(max(len(ls)/lolos,len(tls)/tidak_lolos))

        else:
            supls = len(ls)/lolos
            suptls = len(tls)/tidak_lolos
            growtValueMax = max(supls,suptls)
            growtValueMin = min(supls,suptls)
            GrowthRateValue = growtValueMax/growtValueMin
            grx.append(GrowthRateValue)
            conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        num(dx,n-1,suppdx1,suppdx2,lls,tlls,grx,conf)
    # 501-600
    elif n == 3:
        ls = df[(df[dx]>500) & (df[dx]<600)  & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]>500) & (df[dx]<600)  & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        if len(tls) == 0 or len(ls) == 0:
            grx.append(max(len(ls)/lolos,len(tls)/tidak_lolos))
            conf.append(max(len(ls)/lolos,len(tls)/tidak_lolos))

        else:
            supls = len(ls)/lolos
            suptls = len(tls)/tidak_lolos
            growtValueMax = max(supls,suptls)
            growtValueMin = min(supls,suptls)
            GrowthRateValue = growtValueMax/growtValueMin
            grx.append(GrowthRateValue)
            conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        num(dx,n-1,suppdx1,suppdx2,lls,tlls,grx,conf)
    # 601-700
    elif n == 2:
        ls = df[(df[dx]>600) & (df[dx]<700)  & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]>600) & (df[dx]<700)  & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        if len(tls) == 0 or len(ls) == 0:
            grx.append(max(len(ls)/lolos,len(tls)/tidak_lolos))
            conf.append(max(len(ls)/lolos,len(tls)/tidak_lolos))

        else:
            supls = len(ls)/lolos
            suptls = len(tls)/tidak_lolos
            growtValueMax = max(supls,suptls)
            growtValueMin = min(supls,suptls)
            GrowthRateValue = growtValueMax/growtValueMin
            grx.append(GrowthRateValue)
            conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        num(dx,n-1,suppdx1,suppdx2,lls,tlls,grx,conf)
    # >700
    elif n == 1:
        ls = df[(df[dx]>700) & (df['Ket (lolos/tidak lolos)']=='Lolos')]
        tls = df[(df[dx]>700) & (df['Ket (lolos/tidak lolos)']=='Tidak Lolos')]
        if len(tls) == 0 or len(ls) == 0:
            grx.append(max(len(ls)/lolos,len(tls)/tidak_lolos))
            conf.append(max(len(ls)/lolos,len(tls)/tidak_lolos))

        else:
            supls = len(ls)/lolos
            suptls = len(tls)/tidak_lolos
            growtValueMax = max(supls,suptls)
            growtValueMin = min(supls,suptls)
            GrowthRateValue = growtValueMax/growtValueMin
            grx.append(GrowthRateValue)
            conf.append(growtValueMax/(len(tls)/tidak_lolos+len(ls)/lolos))
        suppdx1.append(len(ls)/lolos)
        suppdx2.append(len(tls)/tidak_lolos)
        lls.append(len(ls))
        tlls.append(len(tls))
        num(dx,n-1,suppdx1,suppdx2,lls,tlls,grx,conf)

def support(dx):
    if df[dx].dtypes != 'object':
        num(dx,n=5,suppdx1=[],suppdx2=[],lls=[],tlls=[],grx=[],conf=[])
    else:
        str(dx)

# masukkan nama kolom yang ingin dihitung di sini
support('Univ')
```
