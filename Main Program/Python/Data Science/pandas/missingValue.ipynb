```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import random


vals = np.random.randn(50)
vals[random.sample([i for i in range(50)], 5)] = np.nan
df = DataFrame(vals.reshape(10,5),columns = ['A','B','C','D','E'])
df
```

```python
# rename columns
df = df.rename(columns={'A':'Z'})
df.head() # berfungi untuk menampilkan 5 baris pertama
```

```python
# berfungsi untuk mengembalikan data seperti di awal
# sebelum diproses
df_backup = df.copy(deep=True)
```

```python
# menghapus(drop) kolom yang mengandung missing values
df = df.dropna(axis='columns')
df.head()
```

```python
# berfungsi untuk mengembalikan data seperti di awal
# saat data dibackup dari variabel df.backup
df = df_backup.copy(deep=True)

# menghapus(drop) baris yang mengandung missing values
df = df.dropna(axis='rows')
df.head()
```

```python
# presentase missing values tiap kolom
df = df_backup.copy(deep=True)
df.isna().mean()
```

```python
# menghapus(drop) kolom yang mengandung missing values
# berdasarkan threshold
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```
