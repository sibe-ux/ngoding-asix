## Studi Kasus : Indeks Kebahagiaan Indonesia
Indonesia pada tahun 2021 berada di peringkat 82 dalam World Happiness Index tahun 2021, data ini mengalami peningkatan dari tahun lalu di mana Indonesia berada di posisi 84. Peringkat yang dipublikasikan oleh the Sustainable Development Solutions Network (SDSN) dan The Center for Sustainable Development, Columbia University, disponsori oleh PBB pada setiap peringatan hari kebahagian dunia yang jatuh setiap 20 Maret. Peringkat ini disusun melalui pendekatan kombinasi dari beberapa indikator seperti GDP per kapita, bantuan sosial, kesehatan, dll. Pada notebook ini akan divisualisasikan perbandingan indeks kebahagiaan Indonesia dibandingkan dengan negara-negara Asia Tenggara (SEA)

### Langkah 1 : Import library yang digunakan
pada tahap 1 dilakukan import library yang digunakan untuk menganalis dan memvisualisasi Data indeks kebahagiaan Indonesia. Library yang digunakan antara lain: (1) pandas; (2)numpy; dan (3) matplotlib; (4) seaborn.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns 
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap
```

### Langkah 2 : Membuat Color Dictionary
Color dictionary digunakan untuk memudahkan kita dalam melakukan mapping pewarnaan plot yang digunakan dalam visualisasi data. 

```python
colors_blue = ["#132C33", "#264D58", '#17869E', '#51C4D3', '#B4DBE9']
colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
colors_red = ["#331313", "#582626", '#9E1717', '#D35151', '#E9B4B4']
colors_mix = ["#17869E", '#264D58', '#179E66', '#D35151', '#E9DAB4', '#E9B4B4', '#D3B651', '#6351D3']
colors_div = ["#132C33", '#17869E', '#DADADA', '#D35151', '#331313']

sns.palplot(colors_blue)
sns.palplot(colors_dark)
sns.palplot(colors_red)
sns.palplot(colors_mix)
sns.palplot(colors_div)
```

```python
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.left'] = False
mpl.rcParams['axes.titlecolor'] = colors_dark[0]
mpl.rcParams['axes.labelcolor'] = colors_dark[0]

# tick
mpl.rcParams['xtick.color'] = colors_dark[0]
mpl.rcParams['ytick.color'] = colors_dark[0]
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12


# legend 
mpl.rcParams['legend.edgecolor'] = colors_dark[0]
```

### Langkah 3 : Load dan tampilkan dataset yang digunakan
Pada notebook ini digunakan 2 dataset yaitu: (1) world-happiness-report-2021.csv; dan (2) world-happiness-report.csv

```python
data_2021 = pd.read_csv("world-happiness-report-2021.csv")
data_past = pd.read_csv("world-happiness-report.csv")
```

### Langkah 4 : Mengelompokkan Negara Asia Tenggara (SEA)
Karena tujuan kita adalah membandingkan tingkat kebahagian Indonesia dibandingkan dengan negara-negara tetangga atau negara-negara asia tenggara, maka kita perlu mengelompokkan negara-negara dengan atribut Regional indicator = Southeast Asia.

```python
SEA = data_2021[data_2021['Regional indicator'] == "Southeast Asia"]['Country name'].to_list()
```

### Langkah 5 : Memvisualisasikan data perbandingan indeks kebahagiaan Indonesia dengan negara-negara SEA
Setelah kita mengelompokkan data negara-negara SEA, maka kita dapat memvisualisasikan perbandingan indeks kebahagiaan Indonesia dengan negara-negara tetangga. Untuk memvisualisasikannya kita dapat menggunakan barchart dengan mengkolaborasikannya dengan beberapa Exploratory Data Analysis yaitu nilai rata-rata (mean).

```python
df = data_2021[data_2021['Country name'].isin(SEA)]
data_2021_top = data_2021.iloc[0:1]
data_2021_bot = data_2021.iloc[-1]
mean_score = data_2021['Ladder score'].mean()
sea_idx = list(df.index + 1)

fig, ax = plt.subplots(figsize=(14, 8))

bars0 = ax.bar(data_2021_top['Country name'], data_2021_top['Ladder score'], color=colors_blue[0], alpha=0.6, edgecolor=colors_dark[0])
bars1 = ax.bar(df['Country name'], df['Ladder score'], color=colors_dark[3], alpha=0.4, edgecolor=colors_dark[0])
bars2 = ax.bar(data_2021_bot['Country name'], data_2021_bot['Ladder score'], color=colors_red[0], alpha=0.6, edgecolor=colors_dark[0])
line  = ax.axhline(mean_score, linestyle='--', color=colors_dark[2])

ax.legend(["Rata-rata", "Paling bahagia", "Negara SEA", "Paling tidak bahagia"], loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=5, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Negara", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Skor Indeks Kebahagian", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

avgl  = ax.text(
    s="Rata-rata\nGlobal: {:.2f}".format(mean_score),
    x=xmax*1.02,
    y=mean_score,
    backgroundcolor=colors_dark[2],
    fontsize=12,
    fontweight='bold',
    color='white'
)

# Indonesia setting
bars1[5].set_alpha(1)
bars1[5].set_color(colors_red[3])
bars1[5].set_edgecolor(colors_dark[0])

for i, bar in enumerate(bars1) : 
    x=bar.get_x()
    y=bar.get_height()
    if i != 5 : 
        ax.text(
            s=f"{sea_idx[i]}th",
            va='center', ha='center', 
            x=x+0.38, y=y/2,
            color=colors_dark[3],
            fontsize=14,
        )
    else : 
        ax.text(
        s=f"{sea_idx[i]}th",
        va='center', ha='center', 
        x=x+0.38, y=y/2,
        color='white',
        fontsize=14,
        fontweight='bold'
    )
        
for i, bar in enumerate(bars0) : 
    x=bar.get_x(),
    y=bar.get_height(),

    ax.text(
        s=f"1st",
        va='center', ha='center', 
        x=x[0]+0.38, y=y[0]/2,
        color="white",
        fontsize=14,
        fontweight='bold',
        alpha=1,
    )
    
for i, bar in enumerate(bars2) : 
    x=bar.get_x(),
    y=bar.get_height(),

    ax.text(
        s="149th",
        va='center', ha='center', 
        x=x[0]+0.38, y=y[0]/2,
        color="white",
        fontsize=14,
        fontweight='bold',
        alpha=1,
    )
        
plt.text(s="Urutan ke-berapa indeks kebahagiaan Indonesia\ndi antara negara-negara tetangga?", ha='left', x=xmin, y=ymax*1.12, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title("Di antara negara-negara Asia Tenggara, Indonesia menempati urutan ke-6 dalam indeks kebahagiaan.\nDalam konteks dunia, Indonesia menempati urutan ke-82, tetapi masih di bawah rata-rata.", loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```

### Langkah 6 : Memvisualisasikan breakdown indikator tingkat kebahagiaan di negara-negara SEA
Untuk melakukan visualisasi terkait dengan breakdown indikator-indikator yang digunakan untuk mengukur tingkat kebahagiaan negara dapat dialkukan dengan menggunakan stacked bar chart.

```python
components = list(data_2021.columns[-7:-1])
df = data_2021[data_2021['Country name'].isin(SEA)]
df = df[["Country name"] + components].set_index("Country name")

fig, ax = plt.subplots(figsize=(14, 8), dpi=75)

bars = df.plot(kind='bar', stacked=True, ax=ax, color=colors_mix[0:7], alpha=0.8, edgecolor=colors_dark[0], rot=0)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Negara", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Skor", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

plt.text(s="Breakdown skor negara SEA", ha='left', x=xmin, y=ymax*1.18, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title("PDB/ GDP per kapita memainkan peran besar dalam peningkatan skor indeks kebahagiaan\nnegara yang paling tidak bahagia memiliki PDB rendah, tetapi tidak selalu menjadi faktor utama\njuga bisa dilihat bahwa Singapura, negara paling bahagia di Asia Tenggara memiliki tingkat korupsi yang rendah", loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```

### Langkah 7 : Memvisualisasikan breakdown indikator Indonesia dengan Pie Chart
Salah satu teknik plotting yang dapat digunakan untuk membreakdown indikator tingkat kebahagiaan di Indonesia adalah dengan menggunakan pie chart.

```python
components = list(data_2021.columns[-7:-1])
df = data_2021[data_2021['Country name'].isin(SEA)]
df = df[["Country name"] + components].set_index("Country name")
df_idn = df.loc['Indonesia']

fig, ax = plt.subplots(figsize=(14, 8), dpi=75)

pie = ax.pie(
    df_idn, 
    colors=colors_mix[0:7],
    wedgeprops=dict(width=0.5, alpha=0.9),
    autopct='%1.0f%%',
    pctdistance=1.12, 
    textprops={
        'fontsize': 12, 
        'color': colors_dark[2],
        'fontweight': 'bold'
    }, 
)

ax.legend(df_idn.index, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

plt.suptitle(t="Breakdown Skor Indonesia", fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title("Kedermawanan di Indonesia berkontribusi relatif banyak terhadap skor kebahagiaan\ndan sayangnya persepsi korupsi hanya berkontribusi​ 1%", fontsize=13, color=colors_dark[2])
plt.tight_layout()
plt.show()
```

### Langkah 8 : Memvisualisasikan perubahan indeks kebahagiaan Indonesia per tahun sejak 2006
Tahapan berikutnya adalah memvisualisasikan grafik perubahan indeks kebahagiaan Indonesia per tahun dengan menggunakan line chart

```python
df = data_past[data_past['Country name'].isin(SEA)].set_index('Country name')
df_idn = df.loc['Indonesia']
df = df.reset_index()
df = df[df['Country name'] != "Indonesia"]
mean_idn = df_idn['Life Ladder'].mean()

fig, ax = plt.subplots(figsize=(14, 8), dpi=75)

line0 = sns.lineplot(data=df, x='year', y='Life Ladder', hue='Country name', alpha=0.2, ax=ax, palette=colors_mix[:9])
line1 = ax.plot(df_idn.year, df_idn['Life Ladder'], alpha=1, marker='o', color=colors_red[3], linewidth=3, label='Indonesia')
line2 = ax.axhline(mean_idn, linestyle='--', alpha=1, color=colors_dark[1])


ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=5, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Negara", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Skor", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

avgl  = ax.text(
    s="Rata-rata\nSkor\nIndonesia: {:.2f}".format(mean_idn),
    x=xmax+0.5,
    y=mean_idn,
    backgroundcolor=colors_dark[2],
    fontsize=12,
    fontweight='bold',
    color='white'
)

plt.text(s="Skor Kebahagiaan Indonesia Sejak 2006", ha='left', x=xmin, y=ymax*1.08, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title("Sepertinya belum ada peningkatan yang signifikan dalam skor kebahagiaan Indonesia selama 14 tahun terakhir.\nTahun dengan skor kebahagiaan tertinggi untuk Indonesia adalah pada tahun 2014", loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```

### Langkah 9 : Memvisualisasikan persebaran bantuan sosial, harapan hidup sehat, dan PDB per kapita
Beberapa indikator yang menjadi indikator tingkat kebahagiaan adalah bantuan sosial (social support), harapan hidup sehat (healthy life expectancy) dan PDB per kapita (GDP per capita). Karena kita membandingkan tiga dimensi variabel maka tidak dapat divisualisasikan dengan menggunakan scatter plot, sehingga teknik visualisasi yang kita gunakan adalah buble chart.

```python
def getSea(row) : 
    if row['Country name'] == "Indonesia" : 
        return "Indonesia"
    elif row['Country name'] in SEA : 
        return "SEA"
    else : 
        return "Other"
```

```python
df = data_2021
df['Countries'] = df.apply(lambda x: getSea(x), axis=1)
meanx=df['Social support'].mean()
meany=df['Healthy life expectancy'].mean()
singx=df[df['Country name'] == 'Singapore']['Social support']
singy=df[df['Country name'] == 'Singapore']['Healthy life expectancy']

fig, ax = plt.subplots(figsize=(18, 8), dpi=75)

sns.scatterplot(
    data=df, 
    x='Social support', 
    y='Healthy life expectancy', 
    size='Logged GDP per capita', 
    ax=ax, sizes=(5, 1000),
    alpha=0.9,
    hue='Countries',
    palette=[colors_dark[4], colors_blue[1], colors_red[2]]
)
linex = ax.axvline(meanx, linestyle='dotted', color=colors_dark[1], alpha=0.8, label='Average')
liney = ax.axhline(meany, linestyle='dotted', color=colors_dark[1], alpha=0.8)
text  = ax.text(
    s="Singapore",
    x=singx-0.013,
    y=singy+1.5,
    color=colors_dark[2]
)


ax.legend(bbox_to_anchor=(1.05, 1), ncol=1, borderpad=1, frameon=False, fontsize=12)
ax.grid(alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Bantuan sosial", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Harapan hidup sehat", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()


plt.text(s="Bantuan sosial, Harapan hidup sehat\ndan PDB per Kapita", ha='left', x=xmin, y=ymax*1.04, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title("Nampaknya Indonesia masih berada di kuadran ketiga dengan 3 negara SEA lainnya masih memiliki ruang untuk perbaikan\nSingapura memiliki skor terbaik di antara negara-negara SEA", loc='left', fontsize=13, color=colors_dark[2])  
plt.tight_layout()
plt.show()
```
